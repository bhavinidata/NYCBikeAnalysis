```python
import os
import pandas as pd
import numpy as np
from functools import reduce
from datetime import datetime
```

```python
def get_filenames(dirname):
    filenames = os.listdir(dirname)
    file_fullpath = [
        os.path.join(dirname, f)
        for f in filenames
    ]
    return file_fullpath
```

```python
filenames2019 = get_filenames("2019CitybikeData")
filenames2018 = get_filenames("2018CitybikeData")
filenames2017 = get_filenames("2017CitybikeData")
filenames2016 = get_filenames("2016CitybikeData")
```

```python
filenames2017
```

```python
def read_saveCSV(filenames, year):
    citibike_dfs = [
    pd.read_csv(f)
    for f in filenames
    ]
    def concat(df1, df2):
        return pd.concat([df1, df2])

    combined_citybikedf = reduce(concat, citibike_dfs)

    newfilename = "citybikeData" + str(year) +".csv"
    combined_citybikedf.to_csv(f"CombinedBikeData/{newfilename}", index=False)
```

```python
read_saveCSV(filenames2016, 2016)

read_saveCSV(filenames2017, 2017)

read_saveCSV(filenames2018, 2018)

read_saveCSV(filenames2019, 2019)
```

#### Remove the data with trip duration greater than 3 hrs (with different start and stop days) and less than a minute

```python
cleanbikeData2017To2018 = cleanbikeData2017To2018.loc[(cleanbikeData2017To2018["tripduration"]<11000) & (cleanbikeData2017To2018["tripduration"]>60) | (cleanbikeData2017To2018["StartDayofWeek"] == cleanbikeData2017To2018["StopDayofWeek"])]
```

```python
# Calculate the distance from start and end latitude and longitude.
def dist_from_coordinates(lat1, lon1, lat2, lon2):
#     R = 6371  # Earth radius in km
    R = 3959  # Earth radius in mile
    #conversion to radians
    d_lat = np.radians(lat2-lat1)
    d_lon = np.radians(lon2-lon1)

    r_lat1 = np.radians(lat1)
    r_lat2 = np.radians(lat2)

    #haversine formula
    a = np.sin(d_lat/2.) **2 + np.cos(r_lat1) * np.cos(r_lat2) * np.sin(d_lon/2.)**2

    haversine = 2 * R * np.arcsin(np.sqrt(a))

    return haversine


new_column = []                    #empty column for distance
for index,row in cleanbikeData2017To2018.iterrows():
    
    lat1 = row['start station latitude'] #first row of location.lat column here
    lon1 = row['start station longitude'] #first row of location.long column here
    lat2 = row['end station latitude'] #second row of location.lat column here
    lon2 = row['end station longitude'] #second row of location.long column here
    value = dist_from_coordinates(lat1, lon1, lat2, lon2)  #get the distance
    new_column.append(value)   #append the empty list with distance values

cleanbikeData2017To2018.insert(4,"Distance(miles)",new_column)  #4 is the index where you want to place your column. Column index starts with 0. "Distance" is the header and new_column are the values in the column.
```
